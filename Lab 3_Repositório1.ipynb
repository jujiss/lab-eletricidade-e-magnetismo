### III - Circuitos em paralelo *III.6 e III.7*



$$ \frac{1}{R_T} = \frac{1}{R_1} + \frac{1}{R_2} + \cdots + \frac{1}{R_n}, \text{como } R_1 = R_2 = R_3 = \cdots = R_n = R$$

$$ \frac{1}{R_t} = \sum_{i=1}^{n} 1\left(\frac{1}{R}\right) = \frac{n}{R} \Rightarrow \boxed{ R_t = \left(\frac{1}{n}\right) R } $$ 



```python
U = 6
R = 1
vRt = []
vI = []
n = []
for i in range(1, 50):
  n.append(i)
  Rt = (1/i) * R
  vRt.append(Rt)
  vI.append(U/Rt)

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(10, 5))
ax1.plot(n, vRt, 'o-', color = "dodgerblue", label = "Resistência total em função do número de lâmpadas.")
ax1.set_ylabel('Resistência total (Ω)', fontsize = 16)
ax2.plot(n, vI, 'o-', color = "deeppink", label = "Corrente do número de lâmpadas.")
ax2.set_ylabel('Corrente elétrica (A)', fontsize = 16)
# fig.suptitle("Resistência total e corrente elétrica de um circuito em relação ao número de lâmpadas de mesma resistência em paralelo.")
fig.supxlabel('Número de lâmpadas de mesma resistência em paralelo')
plt.show()
```
